```python
import csv
import re
from datetime import *
import json
```

```python
def test(route_given, stop_given):
    
    now = datetime.now()
    current_time = now.strftime("%H:%M:%S")
    
    regex = re.compile("(\-[\w]+\-)")
    no_ceros = re.compile("[1-9][0-9]*")
    
    with open('stop_times.csv', encoding="utf8") as read_file:
        read_csv = csv.reader(read_file, delimiter=",")
        
        smallest = timedelta(minutes=5)
        nearest_time = []
            
        for i, row in enumerate(read_csv):
            if i == 0:
                continue
            else:
                route = regex.search(row[0]).group()
                route = route[1:-1]
                stop_no = no_ceros.search(row[3][-6:]).group()
                                
                if route_given == route and stop_given == stop_no:                  
                    stop_time = datetime.strptime(row[1], "%H:%M:%S")
                    now_parsed = datetime.strptime(current_time, "%H:%M:%S")
  
                    if now_parsed < stop_time:
                        diff = stop_time - now
                        if diff < smallest:
                            smallest = diff
                            nearest_time = {route: {"seq": row[4], "dest": row[5]}}
#         print(nearest_time)                  
    read_file.close()   
    return nearest_time
```

```python
def get_searched_stops(self, dict):

        if len(dict.keys()) == 0:
            print("No routes for the stops given")
        else:
            with open('stops.json', encoding="utf8") as read_file:
                read = json.load(read_file)

                stops = []
                for route in dict.keys():
                    for stop in dict[route]:
                        lat = read[stop]["lat"]
                        lon = read[stop]["lon"]
                        stop_name = read[stop]["stop_name"]
                        routes = [key for key in read[stop]["routes"].keys()]

                        stops.append([lat, lon, stop_name, stop, routes])

                read_file.close()

                return stops
```
