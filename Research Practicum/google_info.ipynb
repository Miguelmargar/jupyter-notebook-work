```python
def fuzz(stopA, stopB):
    data = tuple()
    with open('stops.csv', encoding="utf8") as read_file:
        read_csv = csv.reader(read_file, delimiter=",")
        
        oconnell = "O'Connell Bridge"
        
        if oconnell in stopA:
            stopA = oconnell
        if oconnell in stopB:
            stopB = oconnell
        
        countA = 0
        countB = 0
                
        dataA = [()]
        dataB = [()]
            
        for i, row in enumerate(read_csv):
            if i == 0:
                continue
            else:
                stopA_ratio = difflib.SequenceMatcher(None, stopA, row[4]).ratio()
                stopB_ratio = difflib.SequenceMatcher(None, stopB, row[4]).ratio()
                
                if stopA_ratio > countA:
                    countA = stopA_ratio
                    dataA[0] = (row[0], row[2], row[3], row[4])

                if stopB_ratio > countB:
                    countB = stopB_ratio
                    dataB[0] = (row[0], row[2], row[3], row[4])
                    
                
        data += dataA[0]
        data += dataB[0]
                                        
            
    read_file.close()

    return data

fuzz(stopA, stopB)
```
