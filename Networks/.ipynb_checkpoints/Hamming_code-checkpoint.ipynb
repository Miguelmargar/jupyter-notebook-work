```python
bits = input("Please input the bit stream: ")

def hammingEncoder(bits):
    """
    -Main function to get input from user in binary and return the Hamming Code result
    """
    
    def check_binary(bits):
        """
        Function to check that all input is in binary with either a 0 or a 1
        """
        #Check all input to be in binary and count number of digits if so
        good = ["0", "1"]
        count = 0
        
        # Loop through user input
        for i in bits:

            # If each item is either a 0 or a 1
            if i in good:
                # Increase the count by 1
                count += 1
            # If each item is not a 0 or 1 print error message
            else:
                print("Error: Input must be binary")
                count -= 1
                
        return count == len(bits)
        
            
    def get_mod(bits):
        """
        Function that takes the bit stream and returns a list of the locations of the checkbits and the length of the list 
        describes the amount of checkbits needed.
        """
        # Initiate an empty list
        mod = []
        square = 1
        # Loop through the user input
        for i, k in enumerate(bits):
            # If the index of the item is 0, add 1 to the empty list and multiply square by 2
            if i == 0:
                mod.append(1)
                square *= 2

            # If the index of the item mod is 0, multiply square by 2 and append it to list
            elif i % square == 0:
                mod.append(square)
                square *= 2        

            else:
                continue
        # Return the now full list
        return mod
   

    def get_xor(binary):
        """
        Get the input and return xor of the bit stream
        """
        #check the first digit of the input
        result = 0
        if binary[0] == "1":
            result = 1

        # loop through the input and xor it to get final result
        for i in range(len(binary) -1):
            if (result + int(binary[i + 1])) == 1:
                result = 1
            else:
                result = 0

        return str(result)


    
    def get_encode(bits):
        """
        Function that takes user input and list of check bits' position and number to work out the final 
        Hamming code
        """

        # Use function to get the amount of check bits and locations of each check bit
        mod = get_mod(bits)
        print("mod is: ", mod)
        
        # Create a list with the length of the final Hamming Code and fill it with the char "?"
        end = list("?" * (len(mod) + len(bits)))
        print("The length of end is: ", len(end))
        print("end here is: ", end)
        print("bits is: ", bits)
        print()

        # Substitue "?" with an underscore in the new list "end" the index number of each item
        # in the check bits location list 
        for i in mod:
            end[i - 1] = "_"


        # Loop through the user input and for each bit
        for i in bits:
            # Loop through each item in the new list "end" and when encountering "?" substitute "?" with each bit
            for k, j in enumerate(end):
                if end[k] == "?":
                    end[k] = i
                    break
                # If not "?" keep going with the loop
                else:
                    continue
#         print("end is: ", end)
        print()

        # Loop through the list with the check bits locations
        for i, k in enumerate(mod):
            
            # Create a new list
            mylist = ""
            
            # Loop through the range starting at each check bit minus one, ending at the length of the list "end" and skip
            # the amount of the value multiply by 2 of each item in the check bit location list 
            for j in range(k - 1, len(end), k * 2):
                
                # add to the new list the first item even if yet unknown check bit
                mylist += end[j]
                
                # initiate count to 1 for while loop
                count = 1
                # Add the rest of the bits to "mylist"
                while count <= k-1:
                    if j + count < len(end):
                        mylist += (end[j + count])
                    count += 1   
            
            # call get_xor for each list
            check_bit = get_xor(mylist[1:])
            
            # Add the check bit to the final list
            for e, f in enumerate(end):
                if end[e] == "_":
                    end[e] = check_bit
                    break
                else:
                    continue
        
        hamming_code = ""
        for i in end:
            hamming_code += i
            
        
        print()
        print("Final Hamming code is: ", hamming_code)
        
    
    # If input is all binary 
    if check_binary(bits):
        
        # Get the check bits list and pass it to get_encode()
        get_mod(bits)
        
        # Use function to get the Hamming Code
        get_encode(bits)

        
            
hammingEncoder(bits)
    
```
